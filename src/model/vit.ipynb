# Load pre-trained model checkpoint and image processor

```python
model_checkpoint = "google/vit-base-patch16-224-in21k" # pre-trained model from which to fine-tune
batch_size = 32 # batch size for training and evaluation
```

```python
from transformers import AutoImageProcessor

image_processor  = AutoImageProcessor.from_pretrained(model_checkpoint)
image_processor
```

# Data augmentation

```python
# Data transformation techniques
from torchvision.transforms import (
    CenterCrop,
    Compose,
    Normalize,
    RandomHorizontalFlip,
    RandomResizedCrop,
    RandomRotation,
    RandomAffine,
    Resize,
    ToTensor,
)

# Image normalization
normalize = Normalize(mean=image_processor.image_mean, std=image_processor.image_std)

# Image crop
if "height" in image_processor.size:
    size = (image_processor.size["height"], image_processor.size["width"])
    crop_size = size
    max_size = None
elif "shortest_edge" in image_processor.size:
    size = image_processor.size["shortest_edge"]
    crop_size = (size, size)
    max_size = image_processor.size.get("longest_edge")

# Training data augmentation
train_transforms = Compose(
        [
            RandomResizedCrop(crop_size),
            RandomHorizontalFlip(),
            RandomRotation(degrees=(-10, 10)), # Rotate between -10 and 10 degrees
            RandomAffine(degrees=(-10, 10)), # Rotate between -10 and 10 degrees
            ToTensor(),
            normalize,
        ]
    )

def preprocess_train(example_batch):
    """Apply train_transforms across a batch."""
    example_batch["pixel_values"] = [
        train_transforms(image.convert("RGB")) for image in example_batch["image"]
    ]
    return example_batch

# Validation data augmentation
val_transforms = Compose(
        [
            Resize(size),
            CenterCrop(crop_size),
            ToTensor(),
            normalize,
        ]
    )

def preprocess_val(example_batch):
    """Apply val_transforms across a batch."""
    example_batch["pixel_values"] = [val_transforms(image.convert("RGB")) for image in example_batch["image"]]
    return example_batch
```

```python
# Perform data augmentation and assign them to new variables
train_ds = dataset['train'].with_transform(preprocess_train)
val_ds = dataset['validation'].with_transform(preprocess_val)
```

# Load evaluation metric

# Label generation

```python
# Assign label "NORMAL" to value 0, "PNEUMONIA" to value 1
labels = dataset["train"].features["label"].names
label2id, id2label = dict(), dict()
for i, label in enumerate(labels):
    label2id[label] = i
    id2label[i] = label

id2label[1]
```

# Model fine-tuning

**Load pre-trained model**

```python
from transformers import AutoModelForImageClassification, TrainingArguments, Trainer

model = AutoModelForImageClassification.from_pretrained(
    model_checkpoint,
    label2id=label2id,
    id2label=id2label,
    ignore_mismatched_sizes = True,
)
```

**Adjust model's parameters**

```python
model_name = model_checkpoint.split("/")[-1]

args = TrainingArguments(
    f"{model_name}-finetuned",
    remove_unused_columns=False,
    evaluation_strategy = "epoch",
    save_strategy = "epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=batch_size,
    gradient_accumulation_steps=4,
    per_device_eval_batch_size=batch_size,
    num_train_epochs=20,      # Train with 20 epochs
    warmup_ratio=0.1,
    logging_steps=10,
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",     # Use accuracy to judge the best model
    gradient_checkpointing=True,
    weight_decay=0.01,        # Slighly drop weights to avoid overfitting
    save_total_limit=2,
    lr_scheduler_type="cosine",     # Learning-rate scheduler type
    push_to_hub=True,
)
```

**Accuracy computation**

```python
import numpy as np

# Computes accuracy on a batch of predictions
def compute_metrics(eval_pred):
    """eval_pred: tuple (predictions, label_ids)
       predictions (np.array) : model's logits
       label_ids (np.array) : ground-truth labels"""
    predictions = np.argmax(eval_pred.predictions, axis=1)
    return metric.compute(predictions=predictions, references=eval_pred.label_ids)
```

**Batch preprocessing**

```python
import torch

# Process a batch of data and combine individual data samples into a single batch
def collate_fn(examples):
    # Process pixel values from all examples
    pixel_values = torch.stack([example["pixel_values"] for example in examples])

    # Process labels from all examples
    labels = torch.tensor([example["label"] for example in examples])

    return {"pixel_values": pixel_values, "labels": labels}
```

**Adjust training function**

```python
from transformers import AdamW, get_linear_schedule_with_warmup

# Use Adam optimization
optimizer = AdamW(model.parameters(), lr=args.learning_rate)
# How many times the model's parameters are updated during the training process
num_training_steps = len(train_ds) // args.per_device_train_batch_size * args.num_train_epochs
# Learning-rate warmup
lr_scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=int(0.1 * num_training_steps), num_training_steps=num_training_steps)

trainer = Trainer(
    model,
    args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    optimizers=(optimizer, lr_scheduler),
    tokenizer=image_processor,
    compute_metrics=compute_metrics,
    data_collator=collate_fn,
)
```

# Training

```python
train_results = trainer.train()

trainer.save_model()
trainer.log_metrics("train", train_results.metrics)
trainer.save_metrics("train", train_results.metrics)
trainer.save_state()
```

```python
metrics = trainer.evaluate()

trainer.log_metrics("eval", metrics)
trainer.save_metrics("eval", metrics)
```

# Testing

```python
total_samples = len(dataset['test'])
accuracy = 0

for i in dataset['test']:
    encoding = image_processor(i['image'].convert("RGB"), return_tensors="pt")

    # forward pass
    with torch.no_grad():
        outputs = model(**encoding)
        logits = outputs.logits

    predicted_class_idx = logits.argmax(-1).item()

    if predicted_class_idx == i['label']:
        accuracy += 1

print("Accuracy: ", accuracy/total_samples *100, "%")
```
